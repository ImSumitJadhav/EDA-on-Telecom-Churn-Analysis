# **To convert datatypes of some features**

```python
#Convert character(International plan) datatype to integer or bool
df["International plan"]=df["International plan"].map({"No":0,"Yes":1})
df['International plan']=pd.to_numeric(df['International plan'], downcast='integer')

#Convert character(Voice mail plan) datatype to integer or bool
df["Voice mail plan"]=df["Voice mail plan"].map({"No":0,"Yes":1})
df['Voice mail plan']=pd.to_numeric(df['Voice mail plan'], downcast='integer')
```

#**To find out unique values in dataset**

```python
#to find unique values in each feature and count them
unique_Features = []

#use for loop to find unique values
for i in df.columns:
    unique_Features.append([i, df[i].nunique(), df[i].drop_duplicates().values])
pd.DataFrame(unique_Features, columns = ['Features', 'Unique Number', 'Values'])
```

##**Find out Numerical and categorical feature in dataset**

1.Numerical Variables

```python
#numerical features in dataset
numerical_features=[feature for feature in df.columns if df[feature].dtype !="O" and df[feature].dtype !="bool"]
numerical_features
```

2.Categorical Variables

```python
#Categorical features in dataset 
catg_features=[feature for feature in df.columns if df[feature].dtype =="O" or df[feature].dtype =="bool"]
catg_features
```

#<font color='red'>**3.Visualization to find Behaviour of Data**

# **1.Number of churners**

```python
#count of churn and non churn customers
num_churners=df.groupby("Churn")["Churn"].count()
num_churners
```

```python
#give proper figure size
plt.figure(figsize=(6,6))

#plot a pieplot to show percentage of churn and non churn customers
plt.pie(num_churners,radius=1.8,labels=['Loyal','Churn'],autopct="%.0f%%",shadow=True,explode=[0.0,0.2])
```

**Loyal Customers- 86%**

**Churn Customers- 14%**

# **2.Distribution of Numerical variables by density plot**

```python
# create subplots
fig, axs = plt.subplots(figsize=(22,30))

#distance between two plots
plt.subplots_adjust(hspace=1.2)

# loop through the length of discrete features 
#"n" for iterate position of plot
#"m" for select each feature in discrete variable
for n, m in enumerate(numerical_features[:]):
    # add a new subplot iteratively
    ax = plt.subplot(9, 2, n + 1)

    #to plot count of discrete numericals featuers
    sns.distplot(x=df[m], color="b") 

    # chart formatting
    ax.set_title((m+" "+"Vs"+" "+"density").upper())
    ax.set_xlabel(m+" category")

```

**Most of the plot show normal distribution curve.So we can say that,maximum number of features can show normal distributed data.**

# **3.Distribution of Categorical variables by count plot**

```python
#Distribution of Categorical variables
for i in catg_features[:-1]:
  plt.figure(figsize=(20,6))
  sns.countplot(df[i],palette="Paired")
  plt.title(i+" Vs"+" Count of customers in "+i)
  plt.show()
```

**This plot can shows the information about each state contains how many number of customers are present**

# **4.Check outliers in the dataset**

```python
# create subplots
fig, axs = plt.subplots(figsize=(18,50))

#distance between two plots
plt.subplots_adjust(hspace=0.5)

# loop through the length of discrete features 
#"n" for iterate position of plot
#"m" for select each feature in discrete variable
for n, m in enumerate(numerical_features[:]):
    # add a new subplot iteratively
    ax = plt.subplot(9, 3, n + 1)

    #to plot count of discrete numericals featuers
    sns.boxplot(x='Churn', y = m, data=df, ax=ax)

    # chart formatting
    ax.set_title((m+" "+"Vs"+" "+"Outlier").upper())
    ax.set_xlabel(m+" category")
```

**In some cases outliers are less and in some cases number of outliers are more.**

# **5.To check correlation with target variable**
